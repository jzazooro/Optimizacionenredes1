```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

from sklearn.compose import ColumnTransformer

from scipy import stats
import numpy as np
```

```python
# Cargar el csv
file_path = 'cars.csv'
df = pd.read_csv(file_path, sep=';')
df.head()
```

# Limpieza

```python
# Identificar valores faltantes
missing_values = df.isnull().sum()
missing_values
```

```python
# Porcentaje de valores faltantes
missing_percentage = df.isnull().mean() * 100
missing_percentage
```

```python
# Eliminar filas con valores faltantes
df.dropna(inplace=True)
```

```python
# Identificar valores faltantes
missing_percentage = df.isnull().mean() * 100
missing_percentage
```

```python
df['Tiempo'].min() , df['Tiempo'].max()
```

```python
df['Zona _Renta'].unique()
```

```python
# Definir la función para mapear los valores
def map_values(value):
    if value == 'Medio-Bajo':
        return 0.5
    elif value == 'Medio':
        return 1
    elif value == 'Alto':
        return 2
    else:
        return 1
```

```python
# Aplicar la función a la COMBO_ZONA_TIEMPO 'COMBO_ZONA_TIEMPO_objeto' y crear una nueva columna
df['Zona _Renta'] = df['Zona _Renta'].apply(map_values)
```

```python
le = LabelEncoder()

for column in df.columns:
    if df[column].dtype == type(object):
        df[column] = le.fit_transform(df[column])
```

```python
# Calcular la matriz de correlación
corr = df.corr()

# Aplicar un filtro para mostrar solo las correlaciones superiores a 0.6 o inferiores a -0.6
filtered_corr = corr[(corr > 0.4) | (corr < -0.4)]

# Crear el heatmap solo con las correlaciones filtradas
plt.figure(figsize=(10, 8))
plt.title('CORRELACIÓN')
sns.heatmap(filtered_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```

```python
# Normalizar una columna con Min-Max
df['Zona _Renta'] = (df['Zona _Renta'] - df['Zona _Renta'].min()) / (df['Zona _Renta'].max() - df['Zona _Renta'].min())
```

```python
# Normalizar una columna con Min-Max
df['Tiempo'] = (df['Tiempo'] - df['Tiempo'].min()) / (df['Tiempo'].max() - df['Tiempo'].min())
```

```python
df['COMBO_ZONA_TIEMPO'] = 0.44 * df['Zona _Renta'] + 0.93 * df['Tiempo']
df['COMBO_ZONA_TIEMPO'].min(), df['COMBO_ZONA_TIEMPO'].max()
```

```python
# Graficar la distribución con histograma y densidad
plt.figure(figsize=(10, 6))
sns.kdeplot(df['COMBO_ZONA_TIEMPO'], fill=True)  # kde=True agrega la curva gaussiana de densidad
plt.title(f'Distribución de COMBO_ZONA_TIEMPO')
plt.xlabel('COMBO_ZONA_TIEMPO')
plt.ylabel('Frecuencia')
plt.show()
```
